```python
abs_features = []

for batch in range(11):
    with open('./feature_data_test/web_content_features_'+str(batch)+'_abs.pkl', 'rb') as f:
        tmp = pickle.load(f)
        #print(tmp)
        dt = [[tt for tt in t if len(tt) == 10] for t in tmp if len(t) >0 and len([tt for tt in t if len(tt)==10])]
        #print(dat)
        abs_features += [np.mean(np.array(t), axis=0) for t in dt]


print('abstract feature')
#print(abs_features1)
print(abs_features)
print(len(abs_features))

```

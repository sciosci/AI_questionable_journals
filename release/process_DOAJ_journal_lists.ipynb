```python
def get_date(dt):
    try:
        dt = dt.replace("Jan-","January-")
        dt = dt.replace("Feb-","February-")
        dt = dt.replace("Mar-","March-")
        dt = dt.replace("Apr-","April-")
        dt = dt.replace("Jun-","June-")
        dt = dt.replace("Jul-","July-")
        dt = dt.replace("Aug-","August-")
        dt = dt.replace("Sep-","September-")
        dt = dt.replace("Oct-","October-")
        dt = dt.replace("Nov-","November-")
        dt = dt.replace("Dec-","December-")
        
        if int(dt.split('-')[-1]) < 1677:
            return None
        return datetime.strptime(dt, '%d-%B-%Y')
    except:
        return None
```

```python
def get_date2(dt):
    import datetime
    try:
        if ' ' in str(dt):
            tmp = str(dt).split(' ')[0].split('-')
        else:
            tmp = str(dt).split('-')
        year = int(tmp[0])
        month = int(tmp[1])
        day = int(tmp[2])
        
        if year < 1677:
            return None
        return datetime.datetime(year, month, day)
    except:
        return None
```

```python
def reason_filter(r):
    r =str(r)
    if 'no longer open access' in r.lower() or 'no longer oa' in r.lower() or 'journal not open access' in r.lower() or 'requires user registration to access pdfs' in r.lower():
        return 'ok reason'
    if 'ceased' in r.lower() or 'publication suspended' in r.lower() or 'has not published enough articles in last calendar year' in r.lower():
        return 'ok reason'
    if 'dead' in r.lower() or 'inactive' in r.lower() or 'url no longer works' in r.lower() or 'no working website for the journal' in r.lower():
        return 'ok reason'
    if 'conference journal' in r.lower() or 'book reviews only' in r.lower() or 'repository for conference papers' == r.lower():
        return 'ok reason'
    if 'no longer accepting manuscripts' in r.lower() or 'stopped publishing' in r.lower() or 'has not published enough articles this calendar year' in r.lower() or 'not publishing' in r.lower():
        return 'ok reason'
    if 'no articles published' in r.lower() or 'not published since' in r.lower() or 'no issues since' in r.lower() or 'no longer accepting' in r.lower():
        return 'ok reason'
    if 'incorporated into another journal' in r.lower() or 'merged' in r.lower() or 'combined' in r.lower():
        return 'ok reason'
    if "Removed at publisher's request" == r.lower() or 'Editor requested we remove the journal.' == r.lower() or "Does not directly relate to one of the 18 disciplines currently covered by Cabell's." in r:
        return 'ok reason'
    if '"     "                    "          "         "' == r.lower():
        return 'ok reason'
    if 'the issn 20888708 given by editor leads to another journal' == r.lower() or str(np.nan) == r.lower():
        return 'ok reason'
    if str(np.nan) == r.lower() or 'new url' in r.lower():
        return 'ok reason'
    if len(r)<=3 and r not in ['A','P']:
        return 'ok reason'
    return 'interesting reason'
```

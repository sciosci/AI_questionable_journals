```python
for i in range(0, df.shape[0]):
    print(i)
    url = df['URL'].iloc[i]
    try:
        driver.set_page_load_timeout(30)
        driver.get(url)
        driver.set_page_load_timeout(30)
        print(url)
        driver.save_screenshot('./unpaywall_screenshot/' + url.replace("/", "_") + ".png")
    except:
        pass
```

```python
#use this for labeled data
for i in range(0, df.shape[0]):
    print(i)
    url = df['URL'].iloc[i]
    state = df['Reason'].iloc[i]
    try:
        driver.set_page_load_timeout(30)
        driver.get(url)
        driver.set_page_load_timeout(30)
        print(url)
        if state == 'Still in DOAJ index':
            driver.save_screenshot('./screenshots/in_DOAJ/' + url.replace("/", "_") + ".png")
        else:
            driver.save_screenshot('./screenshots/out_DOAJ/' + url.replace("/", "_") + ".png")
    except:
        pass
        
```

```python
def check_website_title(journal_name, saved_dir,url):
    
    saved_path = saved_dir + url.replace('/','_')
    
    if not os.path.isdir(saved_path):
        return 0
    
    files = os.listdir(saved_path)
    match = 0
    if 'home.html' in files:
        with open(saved_path+"/home.html", "r") as f:
            html = f.read()


        soup = BeautifulSoup(html,features="lxml")
        
        for title in soup.find_all('title'):
            if journal_name == title.get_text().strip():
                return 1
            
    return match
            
        
    
```

```python
def simple_title(title):
    #removed ()
    t1 = title.split('(')[0]
    t2 = t1.split(':')[0]
    #check if should deal with - and &
    return t2.strip()
```

```python
def check_website_link(journal_name, saved_dir,url):
    
    saved_path = saved_dir + url.replace('/','_')
    
    if not os.path.isdir(saved_path):
        return 0
    
    files = os.listdir(saved_path)
    match = 0
    if 'home.html' in files:
        with open(saved_path+"/home.html", "r") as f:
            html = f.read()


        soup = BeautifulSoup(html,features="lxml")
        
        for title in soup.find_all():
        
            if 'visit website' in title.get_text().strip().lower():
                return 1
            
    return match
```

```python
def check_website_title_5(journal_name, saved_dir,url):
    
    saved_path = saved_dir + url.replace('/','_')
    
    if not os.path.isdir(saved_path):
        return 0
    
    files = os.listdir(saved_path)
    match = 0
    if 'home.html' in files:
        with open(saved_path+"/home.html", "r") as f:
            html = f.read()


        soup = BeautifulSoup(html,features="lxml")
        
        for title in soup.find_all():
            
            if Levenshtein.distance(journal_name.lower(), title.get_text().strip().lower()) <=2 :
                return 1
            
    return match
```

```python
def check_website_title_7(journal_name, saved_dir,url):
    
    saved_path = saved_dir + url.replace('/','_')
    
    if not os.path.isdir(saved_path):
        return 0
    
    files = os.listdir(saved_path)
    match = 0
    if 'home.html' in files:
        with open(saved_path+"/home.html", "r") as f:
            html = f.read()


        soup = BeautifulSoup(html,features="lxml")
        
        for title in soup.find_all():
            
            if Levenshtein.distance(journal_name.lower(), title.get_text().strip().lower()) <=2 :
                match += 1
                
    if match >3:
        return 1
    else:
        return 0
```
